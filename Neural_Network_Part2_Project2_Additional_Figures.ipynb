```python
#Histogram of Public Shools 

public_schools = education_data_df[education_data_df['school_type'] == 'Public']

plt.figure(figsize=(10, 6))
sns.histplot(data=public_schools, 
             x='avg_test_score_percent', 
             bins=10,  
             color='#2E86AB',  # Blue
             edgecolor='black',
             stat='count')

plt.xlabel('Average Test Score (%)')
plt.ylabel('Number of Schools')
plt.title('Average Test Scores - PUBLIC Schools')
plt.grid(axis='y', alpha=0.3)
```

```python
# Hisotgram of Private School 

private_schools = education_data_df[education_data_df['school_type'] == 'Private']

plt.figure(figsize=(10, 6))
sns.histplot(data=private_schools, 
             x='avg_test_score_percent', 
             bins=10,  
             color='#A23B72',  # Pink/Magenta
             edgecolor='black',
             stat='count')

plt.xlabel('Average Test Score (%)')
plt.ylabel('Number of Schools')
plt.title('Average Test Scores - PRIVATE Schools')
plt.grid(axis='y', alpha=0.3)
```

```python
#Bar Chart


public_high = education_data_df[(education_data_df['school_type'] == 'Public') & 
                           (education_data_df['grade_level'] == 'High')]

plt.figure(figsize=(10, 6))
sns.barplot(data=public_high, 
            x='school_type', 
            y='percent_minority',
            color='#2E86AB',  # Blue
            edgecolor='black',
            linewidth=1.5)

plt.xlabel('School Name')
plt.ylabel('Percent Minority (%)')
plt.title('Percent Minority Students - PUBLIC High Schools')
plt.grid(axis='y', alpha=0.3)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = education_data_df
print(df.shape)
df.head()
```

```python
# Bar plot: average student_teacher_ratio by school_type
sns.barplot(
    data=education_data_df[education_data_df["school_type"].isin(["Private", "Public"])],
    x="school_type",
    y="student_teacher_ratio",
    color="blue",
    edgecolor="black",
    hue = "school_type"
)
plt.show()
```

```python
# Bar plot: average student_teacher_ratio by school_type
sns.barplot(
    data=education_data_df,
    x="school_type",
    y="student_teacher_ratio",
    color="blue",
    edgecolor="black",
    hue = "school_type"
)
plt.show()
```

```python

# Box plot: distribution of student_teacher_ratio by school_type
sns.boxplot(
    data=education_data_df[education_data_df["school_type"].isin(["Private", "Public"])],
    x="school_type",
    y="student_teacher_ratio",
    color="pink"
)
plt.show()
```

```python
numeric_cols = [
    'funding_per_student_usd', 'avg_test_score_percent',
    'student_teacher_ratio', 'percent_low_income',
    'percent_minority', 'internet_access_percent',
    'dropout_rate_percent'
]

corr = df[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    vmin=-1, vmax=1,
    square=True,
    linewidths=0.5,
    ax=ax
)
ax.set_title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

colors = {'Public': '#378ADD', 'Private': '#D4537E'}

filtered_df = df[df['school_type'].isin(['Public', 'Private'])]

for school_type, group in filtered_df.groupby('school_type'):
    ax.scatter(
        group['funding_per_student_usd'],
        group['avg_test_score_percent'],
        label=school_type,
        color=colors[school_type],
        alpha=0.5,
        s=30
    )

ax.set_xlabel('Funding per student (USD)')
ax.set_ylabel('Avg test score (%)')
ax.set_title('Funding vs Test Score: Public vs Private Schools')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

sns.lmplot(
    data=df,
    x="percent_low_income",
    y="funding_per_student_usd",
    hue="school_type",
    col="grade_level",
    height=5,
    aspect=1,
    scatter_kws={"alpha": 0.6}
)

plt.show()
```

```python
state_funding = (
    df.groupby("state", as_index=False)["funding_per_student_usd"]
      .mean()
      .sort_values("funding_per_student_usd", ascending=False)
)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=state_funding,
    x="state",
    y="funding_per_student_usd",
    palette="viridis"
)

plt.xticks(rotation=45, ha="right")
plt.xlabel("State")
plt.ylabel("Average Funding per Student (USD)")
plt.title("Average Funding per Student by State")
plt.tight_layout()
plt.show()
```

```python
# Filter to the two states
ny_mi = df[df["state"].isin(["New York", "Michigan"])]

# Create side-by-side scatterplots with regression lines
g = sns.lmplot(
    data=ny_mi,
    x="percent_minority",
    y="funding_per_student_usd",
    hue="school_type",
    col="state",
    height=5,
    aspect=1,
    scatter_kws={"alpha": 0.6},
    line_kws={"linewidth": 2}
)

g.set_axis_labels("Percent Minority", "Funding per Student (USD)")
g.set_titles("{col_name}")
plt.show()
```
